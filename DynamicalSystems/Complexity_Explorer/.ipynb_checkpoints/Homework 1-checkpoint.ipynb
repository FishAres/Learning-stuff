```julia
using PyPlot
using Statistics 

logmap(x, args...) = args[1] * x * (1 - x)

function iter_map(fun, x, N, args...)
   y = []
    x_ = x
    for i in 1:N
     x_ = fun(x_, args...)
     append!(y, x_)
    end
    return y
end
```

```julia
x1 = 0.2
x2 = 0.200001
r = 2
N = 200
```

```julia
r = 2
t1 = iter_map(logmap, x1, N, r)
t2 = iter_map(logmap, x2, N, r)
plot(abs.(t2-t1), ".")
title("R = $r")
```

```julia
r = 3.4
N = 200
t1 = iter_map(logmap, x1, N, r)
t2 = iter_map(logmap, x2, N, r)
plot(abs.(t2-t1), ".")
title("R = $r")
```

```julia
r = 3.72
t1 = iter_map(logmap, x1, N, r)
t2 = iter_map(logmap, x2, N, r)
plot(abs.(t2-t1), ".")
title("R = $r")
```

```julia
r = 3.4
t1 = iter_map(logmap, x1, 500, r)[end]
t2 = iter_map(logmap, x2, 500, r)[end]
println(abs.(t2-t1))
```

```julia
using Statistics
r = 3.72
N = 500000
t1 = iter_map(logmap, x1, N, r)
t2 = iter_map(logmap, x2, N, r)
println(mean(abs.(t2-t1)))
```

```julia
r = 3.68725
N = 50
plot(iter_map(logmap, x1, N, r), ".")
```
