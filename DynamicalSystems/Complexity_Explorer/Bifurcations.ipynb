```julia
using PyPlot
using Statistics

logmap(x, args...) = args[1] * x * (1 - x)

function iter_map(fun, x, N, args...)
   y = []
    x_ = x
    for i in 1:N
     x_ = fun(x_, args...)
     append!(y, x_)
    end
    return y
end
```

```julia
r = 3
N = 100
x0 = 0.2
traj = iter_map(logmap, x0, N, r);
```

```julia
function bf_diagram(func, x0, num_points, r_slice)
    bf = []
    r_ = []
    N = 500
    for r = 3.84:r_slice:3.8571
        t = iter_map(func, x0, N, r)[end-num_points:end]
        append!(bf, t)
        append!(r_, r*ones(num_points+1))
    end
    plot(r_, bf, ".", markersize=0.04)
end
```

```julia
bf_diagram(logmap, 0.2, 100, 0.0001)
title("toight")
```

Dark areas are called "unstable periodic orbits", where the chaotic attractor is dense.

Apparently a period 3 orbit implies chaos (period 3 is around r=3.85, bottom)

The object is fractal: zooming in it will look like the entire thing.

Most chaotic systems have fractal structure, but definitely not all.
