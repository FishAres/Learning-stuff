```julia
using PyPlot
using Statistics

logmap(x, args...) = args[1] * x * (1 - x)

function iter_map(fun, x, N, args...)
   y = []
    x_ = x
    for i in 1:N
     x_ = fun(x_, args...)
     append!(y, x_)
    end
    return y
end
```

```julia
function bf_diagram(func, x0, num_points, r_slice)
    bf = []
    r_ = []
    N = 500
    for r = 1:r_slice:1.5
        t = iter_map(func, x0, N, r)[end-num_points:end]
        append!(bf, t)
        append!(r_, r*ones(num_points+1))
    end
    plot(r_, bf, ".", markersize=0.04)
end
```

```julia
bf_diagram(logmap, 0.2, 20, 0.0001)
title("toight");
```

## Feigenbaum number

The ratio between points in R where bifurcations occur is constant

$$ \lim_{n -> \inf} \frac{\Delta_n}{\Delta_{n+1}} = 4.66 $$

Where $\Delta_n$ is the width of each pitchfork.

This holds for any 1-D map with a quadratic maximum (also true for sin map and anything that looks like a parabola near its maximum). 

Strogatz thinks this is awesome and calls it a new physical constant.
