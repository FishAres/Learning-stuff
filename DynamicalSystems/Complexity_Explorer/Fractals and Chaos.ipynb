## Most chaotic attractors show fractal state space structure

Slicing the Lorenz attractor cross-wise yields a cantor set (repeating operations, e.g. removing the middle segment of a line).

We need "fractal" dimensions since we can't talk about the "size" of a cantor set (e.g. difference in taking out middle third or middle fifth of a line) with integer dimensions. 

**Capacity dimension**:

$$ d_{cap} = \lim_{\epsilon -> 0} \frac{\log N(\epsilon)}{\log \frac{1}{\epsilon}} $$

where $N(\epsilon)$ is the number of "balls" of a certain size $\epsilon$ you need to cover the space.

In the L1 norm, the "ball" looks like a diamond. For infinity norm, it's a square. It takes $\frac{1}{\epsilon}$ balls to fill a unit segment. 



This means that the dimension of the cantor set is $\frac{\log 2}{\log 3} \approx 0.63 $, which makes sense since it's less than 1 but > 0. 

Chaos and fractals are **not** inextricably linked. But there is something to the connection.
