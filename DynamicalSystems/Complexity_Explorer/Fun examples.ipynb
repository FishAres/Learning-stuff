## Prediction

"Chaos Kabal" at UCSC wanted to predict the path of a roulette ball.

-> The Eudaemonic Pie (or Newtonian Casino).

They tracked each time the ball went by a certain point (e.g. where the dealer was standing). This resulted in unevenly measured data, but this is stil usable (recall the Interspike Interval strategy).

They moved on to make the *prediction company*, later bought by a big Swiss bank. 

Neural nets are good at short-term predictions with nonlinear dynamical systems, but they fail at longer timescales.

Even nearest-neighbor predictions are often better with enough data (i.e. find the most similar point in the trajectory that the system visited in the past, and say that it will do whatever it did then). Lorenz called this the "method of analogues". This is obviously not robust to noise though.

**Computers are dynamical systems**, and often chaotic.



### Controlling chaotic systems

**Feedback control**: 

Pole placement is moving the attractor manifold to keep you in the same point in state space.



This is like drawing an area where you know that, with your current reaction time, you could move to recover the balanced ball, and outside of which the ball would be lost.

A strategy is to exploit the denseness of an attractor to reach the group of trajectories near where you want the system to be, and exploit the denseness of unstable periodic orbits to find one. Once there, you can use local linear control to stay on the UPO.

This is often very useful, but keep in mind you need to wait for the dynamics to reach your desired state - which can be long. 

This can be overcome by using sensitive dependence on initial conditions: use a few nearby initial coordinates.

Cheap orbits: start on the unstable manifold ahead of the desired point in the dynamics, and ride downhill until you're near the desired point. Then, use minute changes to place yourself on the stable manifold. Sometimes you can do this for free (i.e. if your orbit is homoclinic).
