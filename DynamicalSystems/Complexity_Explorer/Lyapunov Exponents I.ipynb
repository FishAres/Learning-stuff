## Dissipation and attractors

In order for attractors to exist, the state space evolution must "dissipate"

Liouville's theorem. 

If there's no dissipation we can have chaos but no chaotic attractors

Chaos is a property of having unstable manifolds(?)

## Lyapunov Exponents



This is an image of the **static dynamics** of the system on that point. Eu, Es correspond to the eigenvectors. The curves derive from running time forward using those eigenvectors.

There are as many Lyapunov exponents $\lambda$ as there are dimensions in the system. Negative exponents compress state space along stable manifolds - positive exponents *stretch* it along unstable manifolds.

The sum of the lyapunov exponents needs to be negative if you want dissipation. 

Chaotic systems have at least one positive exponent!

The biggest positive exponent will win at the limit

The distance between two initially nearby trajectories is 

$$ |\bar\delta_i| \propto e^{\lambda_1 t} $$

Lyapunov exponents contain re-normalization, so you're essentially only measuring locally. 

If you have equations, you can calculate the Lyapunov exponents by getting the eigenvalues of the *variational matrix*

**Local exponents**: In smaller dt's. It's about capturing regional changes in the "shape" of an attractor.
