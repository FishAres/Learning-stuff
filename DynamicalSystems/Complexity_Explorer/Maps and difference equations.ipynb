## Maps

Maps are defined in *discrete* time. This is significant because the mathematics are considerably easier. The next state is fully determined by the previous one.

Example: logistic map  - $x_{n+1} = Rx_n (1 - x_n)$

```julia
logmap(x, args...) = args[1] * x * (1 - x)

function iter_map(fun, x, N, args...)
   y = []
    x_ = x
    for i in 1:N
     x_ = fun(x_, args...)
     append!(y, x_)
    end
    return y
end
```

```julia
using PyPlot

figure(figsize=(5, 3))
plot(iter_map(logmap, 0.2, 20, 2.2));
```

For the value $R=2$ we can see that the fixed point is ~ 0.5

## Transients and attractors

### Attractors (Fixed Points)

Points where $f(\hat x) = \hat x $

This doesn't tell you what the stability of the fixed point is

```julia
plot(iter_map(logmap, 0.2, 50, 2.7), "."; label="R=2.7")
plot(iter_map(logmap, 0.2, 50, 2.8), "."; label="R=2.8")
legend()
title("x0 = 0.2")
```

### Periodic orbit 

This is the case in the logistic map when R=3

```julia
R = 3
plot(iter_map(logmap, 0.2, 50, R), "."; label="R=$R")
R2 = 2.8
plot(iter_map(logmap, 0.2, 50, R2), "."; label="R=$R2")
legend()
title("Period = 2")
```

```julia
R2 = 3.0
plot(iter_map(logmap, 0.2, 50, R2), "."; label="R=$R2")
R2 = 3.5
plot(iter_map(logmap, 0.2, 50, R2), "."; label="R=$R2")
legend(loc="center")
title("4-Cycle")
```

## The changes with different R are called *bifurcations*

```julia
R2 = 3.6
plot(iter_map(logmap, 0.2, 200, R2), "."; label="R=$R2")
# legend(loc="center")
title("Chaotic Attractor")
```
