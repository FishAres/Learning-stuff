### Chaotic data is highly sensitive to frequency filtering

It is not often the case that chaotic data are limited to a certain frequency band.

There are, however, ways to de-noise systems.





### Farner/Sidorowich filtering

Moving time forward will give you the **points moved on the unstable manifold**, while running them backward will give **points moved on the stable manifold**. The true value will be somewhere in the overlap region, which is smaller than the original "noise ball". The manifolds need to be transverse (i.e. they cross). 

To do this time running, you'll need some kind of solver (like 4th order RK) or a model of the system. 

There are ways to de-noise signals by using the *topology* of the attractor (in this case the "connectedness". Noise adds "isolated points" to the set, which can be detected using computational topology techniques and can reasonably safely be thrown away.

*Reference*: Robins et al., Intelligent Data Analysis - Chaos

**Continuity**: If two nearby points wind up in very different places after one timestep, then there may have been very different dynamics working on them. This implies the presence of regime shifts, like bifurcations or noise.

## Getting instantaneous rates

### Reynolds decomposition:

In the case of coral eggs fertilized by sperm, the rate of fertilization is proportional to the product of the concentrations of the two:

$$\text{Rate} = \phi C_E C_S $$

To get instantaneous rates, we can decompose each concentration to the sum of the mean (time average) concentration and the fluctuating component.

$$ C_E = \bar C_E + c'_E(t) $$

$$ C_S = \bar C_S + c'_S(t) $$

Then, with substitution:

$$ \text{Rate} = \phi \bar C_E \bar C_S + \phi c'_E c'_S $$

The second term is the correlation between the fluctuating quantities



This leads to significantly higher aggregation than predicted by the average.

### Lagrangian Coherent Structures (LCS)

Way to understand structures that exist in flow fields like eddies and vortices, generally regions in the system that act as attractors. 

The LCS' are invariant given a flow. That means that different initial points in the flow will always converge on the LCS' despite following very different trajectories. This means you don't need to know the initial conditions! (as long as certain conditions are met).

There are special cases where the trajectories remain separated with certain initial conditions, and ones where you can demix the trajectories. 

In this egg-sperm case, things are made more interesting by the fact that the filaments are made of sperm and eggs, and sperm follow very complex chemotactic gradients (often trytptophan) to find the eggs.
