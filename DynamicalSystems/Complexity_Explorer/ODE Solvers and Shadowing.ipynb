## The error in the Forward Euler method comes from the fact that it's basically the first two terms of the Taylor series for estimating the next point

$$f(x + \Delta x) = f(x_0) + \Delta x (f'(x_0)) + \frac{1}{2}(\Delta x)^2 (f''(x_0)) + .. $$ 

We can call this the "truncation error", i.e. from terms in teh series you don't use. 

Dynamical errors: can snowball when they're fed into the next iteration of the system. 

The Euler method has an error (approximately) that scales with the square of the step size:

$$\frac{1}{2}(\Delta x)^2 (f''(x_0))$$

And is thus an order 2 (or $O(\Delta x^2)$) error.

The *trapezoidal* method uses the average of the Euler and Backward Euler methods and is order $h^3$

## Adaptive timestep ODE solvers do what their name suggests

You can take e.g. two smaller $\frac{h}{2}$ steps and see if the error is within some user-specified tolerance wrt h. If that error is too big, you can start decreasing your stepsize. Likewise, you can increase it when you see the dynamics getting "slower" again. 

Timestep differences can cause bifurcations and often look like "real" physical effects. So investigating potential spurious numerical dynamics is paramount. But beware of machine numerical error ("machine epsilon").



## Shadowing lemma

If you add noise to a trajectory on a chaotic attractor, you'll land on a "shadow trajectory". 

I.e. noise bounces you to an attractor you'd have gotten to eventually. Does not hold if there's noise in the parameters or if the noise is too big (i.e. bounces you out of the basin of attraction)

With good solvers this still happens but on tiny scales



On limit cycles, you'll step off the limit cycle but converge back to it. 

The lemma requires hyperbolicity, i.e. the presence of both a stable and unstable manifold.
