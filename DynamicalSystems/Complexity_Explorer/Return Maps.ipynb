## Correlation / Cobweb map

plot $x_n$ vs $x_{n+1}$

```julia
using PyPlot
using Statistics

logmap(x, args...) = args[1] * x * (1 - x)

function iter_map(fun, x, N, args...)
   y = []
    x_ = x
    for i in 1:N
     x_ = fun(x_, args...)
     append!(y, x_)
    end
    return y
end
```

```julia
x = 0:0.01:1
r = 3.4
plot(x,logmap.(x, 2)) # parabola of the function
plot(x,x)
```

The fixed points f(x) = x lie on the diagonal line
