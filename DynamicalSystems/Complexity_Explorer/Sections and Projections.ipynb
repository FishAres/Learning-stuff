## Unstable periodic orbits

There are an infinite number of unstable periodic orbits in any chaotic attractor. They are also of all periods.

3D is a necessary condition for chaos in flows (but not in maps).

UPOs are dense in a chaotic attractor. You will catch a ride on one no matter how small your step size / object is. 

Attractor is the closure of the set of unstable period orbits. If you got every UPO and "stacked" them, you'd get the original attractor. A fun way to think of them is as compositional "blocks" of the attractor. 

Any trajectory will visit every UPO, a number of times proportional to its stability. You never get "on it", you're just near it. 




This map has a certain periodicity and it re-visits its earlier state.

Poincare surfaces of UPOs: can learn about the periodicity of orbits by the adjacency of nearby points (people use clustering). 

What you really see when you plot a chaotic attractor is an amalgamation of multiple nearby periodic orbits, since every step will get you off of any given orbit.
