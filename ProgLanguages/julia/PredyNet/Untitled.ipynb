```julia
using Plots
using Distributions
using AutoGrad
using Flux
```

```julia
timelength = 1000;
x = linspace(-4, 4, timelength);
y = sin.(x);
```

```julia
W = (randn());
V = (randn());
y_pred = randn();
err = 0;
e_prev = randn();
y_prev = 0;
lr = 0.1
ϵ = 1e-6
```

```julia
Err = zeros(timelength);
Y = zeros(timelength);
lr = 0.01;
function loss(x,y)
    return ((x-y) / (y+x))
end
```

```julia
gradfun = grad(loss)
```

```julia
for t = 1:timelength
    err = (y[t] - y_prev*V);
    y_pred = σ.(e_prev*W);

    g = gradfun(y[t], y_prev*V)
    W -= lr*(g + ϵ)
    V -= lr*(g + ϵ)
    
    y_prev = y_pred
    e_prev = err
    
    Err[t] = err
    Y[t] = y_pred
end
```

```julia
plot(x,y)
plot!(x,Err)
```
