```julia
using Turing
using Mamba: plot, describe
using Gadfly
using Distributions
```

```julia
σ(x) = (1/(1+exp(-x)))
```

```julia
@model gmod(x) = begin
    A ~ Poisson(4)
    B ~ Poisson(0.6)
    x[1] ~ Poisson((A+B))
    return A, B
end
```

```julia
c = sample(gmod([1]), PG(50,300));
```

```julia
p = plot(c); Mamba.draw(p)
```
